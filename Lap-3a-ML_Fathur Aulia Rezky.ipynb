## Decision tree

```python
#import data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('tennis.csv')
data
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
for col in ['outlook', 'temp', 'humidity', 'windy', 'play']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    data
    print (data['outlook'],data['temp'],data['humidity'],data['windy'],data['play'])
```

```python
data
```

```python
# Menentukan data x dan y
x = data.drop('play', axis = 1)
y = data['play']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state =0)
```

## Model

```python
from sklearn.tree import DecisionTreeClassifier
model_DT = DecisionTreeClassifier()
model_DT.fit(x_train, y_train)
```

```python
predict_DT = model_DT.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print ("akurasi scorenya sebesar: ", accuracy_score(y_test, predict_DT)*100)
```

```python
prediksi_baru_DT = model_DT.predict(x_baru_DT)
print('prediksi data : ', ('Yes' if prediksi_baru_DT == 1 else 'No'))
```
